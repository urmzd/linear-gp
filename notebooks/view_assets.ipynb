```python

from pathlib import Path
from IPython.display import display
from IPython.display import Image


# load images from assets/images (**/*.png)
def load_and_show_images(path="../assets/images", ext="png"):

    path = Path(path)
    paths = list(path.glob(f"*.{ext}"))
    paths.sort()

    for file in paths:
        display(Image(file))

load_and_show_images()



```
